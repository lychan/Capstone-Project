Now that I have finished taking a quick look into the columns with mixed dtypes, I want to look at all the dataframes to make sure that they all have the same amount of columns

```python
def multidf_shape(df):
    """prints out the shape of multiple dataframes at once"""

    for i, df in enumerate(df[0:],0): #i= whichever value the function is indexing. df[0:] slices the df at row 1, going from list index 0 
        print(df.shape) #print out the shape of each dataframes

multidf_shape([df19,df20,df21,df22,df23])
```

Next, I want to look at the headers to make sure that all 5 dataframes have the same headers

```python
def compare_headers(df): # a function that took me forever to write, but worth it
    """Compares headers across multiple dataframes against base header and print out 'Columns Unmatched' if there is difference
    and 'Columns Match' if there are the same."""

    header_0 = df[0].columns #this takes the header of the first dataframe in the list, which is index 0, and compares the headers of the subsequent dataframes in the list against it
    for i, df in enumerate(df[0:], 1): # the enumerate function reiterates down the list in row 1, starting at index 1
        diff = set(header_0) != set(df.columns) # making headers into a set to compare. Defining the variable "diff". If the sets don't equal, then they are different and gets stored as variable "diff". 
        if diff: # if the comparison between current index and the base index (index 0) is different
            print(f"Headers Unmatched") # if the above condition is met, print "Headers Unmatched"
        else:
            print(f"Headers Match")  # if the above condition is not met, print "Headers Matched"

compare_headers([df19,df20,df21,df22,df23])
```

By logic, if header 1 = header 0, and all subsequent header all = header 0, then all headers in all dfs match each other. Since they are all the same. We can concatenant into a single master file/dataframe to work with
